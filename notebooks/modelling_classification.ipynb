# Logistic Regression With N-grams = 1

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = utils.load_transformed_data('bag_of_words_three_class')
```

```python
c_range = [
    (1/10)**n
    for n in np.arange(0,10,1)
]

train_accuracy = []
valid_accuracy = []

for c in c_range:
    log_reg = LogisticRegression(penalty='l2', C=c, solver='newton-cholesky')
    log_reg.fit(X_train, y_train)
    train_accuracy.append(log_reg.score(X_train, y_train))
    valid_accuracy.append(log_reg.score(X_val, y_val))
```

```python

plt.figure()
plt.title('Effect on Model Accuracy by Varying C')
plt.plot(c_range, train_accuracy, label='Train')
plt.plot(c_range, valid_accuracy, label='Validation')
plt.xscale('log')
plt.xlabel('log(C)')
plt.ylabel('Accuracy')
plt.axvline(x=c_range[np.argmax(valid_accuracy)], c='red', linestyle='--')
plt.ylim(0.2,1)
plt.legend()
plt.show()


print(f'Max Accuracy Validation: {valid_accuracy[np.argmax(valid_accuracy)]}')
print(f'C: {c_range[np.argmax(valid_accuracy)]}')

print(f'Accuracy of Train: {train_accuracy[np.argmax(train_accuracy)]}')
```

```python
log_reg = LogisticRegression(
    penalty='l2', C=c_range[np.argmax(valid_accuracy)], solver='newton-cholesky'
    )
log_reg.fit(X_train, y_train)

print(f'Test Accuracy: {log_reg.score(X_test, y_test)}')
```

# Random Forest for Classification n-grams = 1, 2 and 3

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = utils.load_transformed_data('3_grams_three_class')
```

```python
max_depth_range = np.arange(1,11,1)

train_accuracy = []
valid_accuracy = []

for max_depth in max_depth_range:
    random_forest = RandomForestClassifier(
        n_estimators=1000,
        max_depth = max_depth,
        n_jobs = -1,
        criterion='entropy',
        max_features='log2'
    )

    random_forest.fit(X_train, y_train)

    train_accuracy.append(random_forest.score(X_train, y_train))
    valid_accuracy.append(random_forest.score(X_val, y_val))
```

```python

plt.figure()
plt.title('Effect on Model Accuracy by Varying Max Depth')
plt.plot(max_depth_range, train_accuracy, label='Train')
plt.plot(max_depth_range, valid_accuracy, label='Validation')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.axvline(x=max_depth_range[np.argmax(valid_accuracy)], c='red', linestyle='--')
plt.ylim(0.2,1)
plt.legend()
plt.show()


print(f'Max Accuracy Validation: {valid_accuracy[np.argmax(valid_accuracy)]}')
print(f'Max Depth: {max_depth_range[np.argmax(valid_accuracy)]}')

print(f'Accuracy of Train: {train_accuracy[np.argmax(train_accuracy)]}')
```

```python
pd.DataFrame(
    {
        'Max Depth': max_depth_range,
        'Validation Accuracy': valid_accuracy,
        'Train Accuracy': train_accuracy
    }
).sort_values(by='Validation Accuracy', ascending=False)
```
